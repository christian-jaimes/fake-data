```python
import pandas as pd
import random
import numpy as np
from google.colab import files
from iso3166 import countries
```

```python
number_of_countries = 4
sexes = ['Male', 'Female']
races = ['White', 'Black', 'Asian', 'Hispanic', 'Native American', 'European']
years = list(range(2014, 2024))
ages = list(range(0, 99))

country_list = [country.name for country in countries][:number_of_countries]

num_rows = 30000  # Number of rows to generate
data = {
    'Origin': [random.choice(country_list) for _ in range(num_rows)],
    'Sex': [random.choice(sexes) for _ in range(num_rows)],
    'Race': [random.choice(races) for _ in range(num_rows)],
    'age': [random.choice(ages) for _ in range(num_rows)],
    'Quantity': [random.randint(0, 100000) for _ in range(num_rows)],
    'Year': [random.choice(years) for _ in range(num_rows)]
}

pop_est = pd.DataFrame(data)

pop_est.head()
```

```python
output_file = 'population_estimates.csv'
pop_est.to_csv(output_file, header=True, index=False, encoding='utf-8-sig')

# Download the file to your local machine
files.download(output_file)
```
