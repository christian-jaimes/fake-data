```python
import pandas as pd
import uuid
from faker import Faker
from google.colab import files
```

```python
# Initialize Faker instance
fake = Faker()

# Define the number of leads you want to generate
num_leads = 2746

# File name to save the CSV data
output_file = "salesforce_leads.csv"

```

```python
# Function to generate unique IDs and emails
def generate_leads(num_leads):
    leads = []

    for _ in range(num_leads):
        unique_id = str(uuid.uuid4())
        first_name = fake.first_name()
        last_name = fake.last_name()
        position = fake.job()
        gender = fake.random_element(["Male", "Female", "Other"])
        company = fake.company()
        email = fake.unique.email()
        website = fake.url()
        salutation = fake.prefix()

        leads.append({
            "Unique ID": unique_id,
            "First Name": first_name,
            "Last Name": last_name,
            "Position": position,
            "Gender": gender,
            "Company": company,
            "Email": email,
            "Website": website,
            "Salutation": salutation
        })

    return leads
```

```python
# Generate leads data
leads = generate_leads(num_leads)

df = pd.DataFrame(leads)
df
```

```python
# Save DataFrame to CSV
output_file = "salesforce_leads.csv"
df.to_csv(output_file, index=False, encoding="utf-8")

# Download the file in Google Colab
files.download(output_file)

print(f"Successfully generated {num_leads} leads and prepared '{output_file}' for download.")
```
